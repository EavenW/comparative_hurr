```python
rotation = 360
number_of_spins = 0

wind_dir_total = np.ones_like(ndb['WDIR'])
wind_dir_total[0] = ndb['WDIR'][0]
for i in range(len(ndb)-1):
    j = i+1
    if ndb['WDIR'][j-1] > 260 and ndb['WDIR'][j] < 100:
        number_of_spins += 1
    if ndb['WDIR'][j-1] < 100 and ndb['WDIR'][j] > 260:
        number_of_spins -= 1
    wind_dir_total[j] = ndb['WDIR'][j] + rotation*number_of_spins

dtheta = np.gradient(wind_dir_total)
```

```python
number_of_spins = 0
wind_dir_total_65 = np.ones_like(d65['WDIR'])
wind_dir_total_65[0] = d65['WDIR'][0]
for i in range(len(d65)-1):
    j = i+1
    if d65['WDIR'][j-1] > 260 and d65['WDIR'][j] < 100:
        number_of_spins += 1
    if d65['WDIR'][j-1] < 100 and d65['WDIR'][j] > 260:
        number_of_spins -= 1
    wind_dir_total_65[j] = d65['WDIR'][j] + rotation*number_of_spins

dtheta_65 = np.gradient(wind_dir_total_65)
```

```python
number_of_spins = 0
wind_dir_total_tck = np.ones_like(tck['direction'])
wind_dir_total_tck[0] = tck['direction'][0]
for i in range(len(tck)-1):
    j = i+1
    if tck['direction'][j-1] > 260 and tck['direction'][j] < 100:
        number_of_spins += 1
    if tck['direction'][j-1] < 100 and tck['direction'][j] > 260:
        number_of_spins -= 1
    wind_dir_total_tck[j] = tck['direction'][j] + rotation*number_of_spins

    
dtheta_tck = np.gradient(wind_dir_total_tck)
```

```python
import scipy.interpolate as interp

eyepassage = mdates.date2num(datetime.datetime(2011,8,28,9))
ru16_datenums = mdates.date2num(ru16.time)
deltaT = np.abs(ru16_datenums - eyepassage)
k_eye = np.where(deltaT == np.nanmin(deltaT))[0][0]
```

```python
profile_tmp = ru16.temperature[k_eye,:].values
d           = ru16.depth.values

def pycnocline_width(profile_tmp, d, low=10.5, high=11.5):
    finite = np.where(np.isfinite(profile_tmp))[0]
    maxdepth = d[np.min(finite)]
    mindepth = d[np.max(finite)]
    new_d = np.arange(mindepth,maxdepth,0.1)
    y = interp.interp1d(d[finite],profile_tmp[finite], kind='linear')
    new_tmp = y(new_d)
    tmp_low = np.abs(new_tmp - low)
    tmp_high = np.abs(new_tmp - high)
    q_low = np.where(tmp_low == np.nanmin(tmp_low))[0][0]
    q_high = np.where(tmp_high == np.nanmin(tmp_high))[0][0]
    #delta_d = new_d[q_high] - new_d[q_low]
    return new_d[q_high], new_d[q_low]

#print(delta_d)
#plt.plot(ru16.temperature[k_eye,:], ru16.depth)
#plt.plot(y(new_d),new_d)
```

```python
profile_tmp = ru16.temperature[k_eye-10,:].values

low_depth  = np.ones_like(ru16['mdate'].values)
high_depth = np.ones_like(ru16['mdate'].values)

for I in range(len(ru16['mdate'].values)):
    profile_tmp = ru16.temperature[I,:].values
    a, b = pycnocline_width(profile_tmp, d, low=14, high=17)
    if np.abs(a-b) < 0.25:
        low_depth[I], high_depth[I] = np.nan, np.nan
    else:
        low_depth[I], high_depth[I] = a, b
```

```python
### find the work
plt.plot(ru16.longitude, ru16.latitude)

def min_distance(lons,lats,glon,glat):
    dist = np.sqrt((glon-lons)**2+(glat-lats)**2)
    a,b = np.where(dist==np.nanmin(dist))
    return a[0], b[0]

def min_dist_linear(lon,lat,glon,glat):
    distlon = np.abs(lon-glon)
    klon = np.where(distlon == np.nanmin(distlon))[0]
    distlat = np.abs(lat-glat)
    klat = np.where(distlat == np.nanmin(distlat))[0]
    return klon,klat
    
```

```python
lons, lats = np.meshgrid(df.Longitude.values, df.Latitude.values)
```

```python
tlen = df.time.shape[0]
klon = np.ones(tlen)
klat = np.ones(tlen)

for i in range(tlen):
    klon[i], klat[i] = min_dist_linear(df.Longitude.values, df.Latitude.values,ru16.longitude.values[i],ru16.latitude.values[i])
```

```python
z_avg = np.ones(tlen)
z_std = np.ones(tlen)
z_all = np.ones((tlen,9))
```

```python
for i in range(tlen):
    r = klat[i]
    s = klon[i]
    loc      = np.meshgrid(np.arange(r-1,r+2),np.arange(s-1,s+2))
    loclat   = loc[0].reshape(9)
    loclon   = loc[1].reshape(9)
    z = np.ones(9)
    for j in range(9): 
        z_all[i,j] = mldroll[int(loclat[j]),int(loclon[j]),i]
```

```python
dTdz = np.abs(np.gradient(ru16.temperature,axis=1))
plt.pcolor(MDATE, D, dTdz)
plt.colorbar()
plt.ylim([-50,0])
print(dTdz.shape)
```

```python
DEPTH = np.abs(ru16.depth)
i = 100
dTdz[i,:]
float(DEPTH[np.where(dTdz[i,:] == np.nanmax(dTdz[i,:]))[0][0]])
```

```python
TLEN = len(ru16.mdate)
GLIDER_MLD = np.ones(TLEN)
for i in range(TLEN):
    try:
        mixed_layer = float(DEPTH[np.where(dTdz[i,:] == np.nanmax(dTdz[i,:]))[0][0]])
        
        if mixed_layer > 5: # and np.nanmax(dTdz[i,:]) > 0.3:
            GLIDER_MLD[i] = mixed_layer
        else:
            GLIDER_MLD[i] = np.nan
    except: GLIDER_MLD[i] = np.nan
```

```python
plt.scatter(ru16.mdate,GLIDER_MLD)
```

```python
rmax = (df.MdatesTimenum.values[1]-df.MdatesTimenum.values[0])
```

```python
def rolling_av(g_time,m_time,g_mld,rmax,k):
    r = np.abs((g_time-m_time[k])/rmax);
    Q = np.where(np.greater_equal(1,r));
    M = g_mld[Q]
    return np.nanmean(M), np.nanstd(M)
```

```python
g_MLD_avg = np.ones(tlen)
g_MLD_std = np.ones(tlen)

for i in range(tlen):
    a, s = rolling_av(ru16.mdate.values,df.MdatesTimenum.values,GLIDER_MLD,rmax,i)
    g_MLD_avg[i] = a
    g_MLD_std[i] = s
```
